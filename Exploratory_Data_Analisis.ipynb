### Persebaran Data 

Pada tahapan ini akan dilihat persebaran data dari kolom populasi dengan nilai kelas yang saya hitung dengan rumus **Freedman-Diaconis rule**. <br>
Freedman-Diaconis rule memanfaatkan 2 rumus dengan **berat bin** dan **bins itu sendiri**. Bin width memanfaatkan nilai q3 dan q1, sedangkan nilai bins itu sendiri memanfaatkan nilai max dan nilai serta berat bin. Berikut formulanya.
![This in an image](https://drive.google.com/uc?id=1xl4pAgJGpWQYwqBHBdgOe0F7cHlsyztu)

Disni saya akan menghitung nilai q1 dan q3 menggunakan numpy dengan fungsi quantile 0.25(25%) dan 0.75 (75%)

### Kolom Population 

```python
fig = plt.figure(figsize=(15,20))
bins_width = (2 * (df.population.quantile(0.75) - df.population.quantile(0.25))) / (len(df.population) ** (1/3))
bins_self = int(np.ceil((df.population.max() - df.population.min()) / bins_width))
sns.histplot(data = df, x = 'population', kde = True, bins = bins_self)
plt.title('Persebaran Populasi Dengan Kelas {}'.format(bins_self), fontsize = 25, pad = 15)
plt.xlabel('Populasi', fontsize = 15, labelpad = 15)
plt.ylabel('Frekuensi', fontsize = 15, labelpad = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.savefig('persebaran_populasi.png')
plt.show()
```

### Cek Normalitas 

Sebelumnya saya telah membuat hipotesis dengan ketentuan H0 dengan nilai signifikan 0.05
1. Populasi, life_exp dan GDP bersifat distribusi normal

Sehingga Jika P-Value < 0.05 maka H0 ditolak.

```python
kolom_popu = kstest(df['population'],"norm")
kolom_harapan = kstest(df['life_exp'], 'norm')
kolom_gdp = kstest(df['gdp_cap'], 'norm')

data = { 'kolom' : ['Populias', 'Harapan', 'GDP'],
          'P-Value' : [kolom_popu[1], kolom_harapan[1], kolom_gdp [1]]
       }

hasil = pd.DataFrame(data = data)
hasil['keputusan'] = ['Hipotesis Nol Ditolak' if x < 0.05 else 'Hipotesis Nol Diterima' for x in hasil['P-Value']]
```

### Mengetahui seberapa besar korelasi antara variabel numerik

Pada tahapan ini, akan mencari dicari korelasi antar kolom numerik. Tujuannya untuk mengetahui sebarap kuat hubungan antar kolom tersebut dengan patokan kolom yang digunakan adalah GDP dikarenakan kolom GDP akan dijadikan label (Y) untuk membuat model machine learning tersebut. Pada tahapan ini akan dilakukan dua tahapan
1. Mencari korelasi sebelum outlier dihapus
2. Mencari korelasi setelah outlier dihapus (Penghapus outlier akan menggunakan metode IQR dengan mencari terlebih dahulu nilai upper whisker dan lower whisker)

Korelasi yang digunakan adalah **Rank Spearman** dikarenakan H0 yang mengasumsikan data berdistribusi normal ditolak. Formula Rank Spearman seperti berikut
![spearman](https://drive.google.com/uc?id=1MbU0WKUH0obm00ccMpQQHlrdF0QrLWhO)

```python
def spearman_correlation(df):
    plt.figure(figsize=(15,10))
    sns.heatmap(df.corr(method = 'spearman'), annot = True, annot_kws = {'fontsize' : 15}, linewidth = 0.1)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
```

#### Korelasi sebelum outlier dihapus 

```python
spearman_correlation(df)
```

#### Korelasi setelah outlier dihapus

Tahapan yang dilakukan
1. Melakukan visualisasi dengan boxplot
2. Mencari nilai IQR
3. Mencari nilai Upper dan Lower Whisker
4. Menghapus outlier yang berada di atas Upper Whisker dan di bawah Lower Whisker

```python
fig, axes = plt.subplots(1,3, figsize=(15,5))

for i, ax, j in zip(df[['population', 'life_exp', 'gdp_cap']], axes.flatten(), range(0,3)):
    sns.boxplot(x = df[i], ax = ax, orient = 'h')
    axes[j].tick_params(axis = 'x', labelsize = 15)
    axes[j].set_xlabel(i, fontsize = 15, labelpad = 15)
    plt.tight_layout()
```

Hasil di atas meninformasikan bahwasanya
1. populasi dan GDP memiliki jumlah outlier paling banyak
2. Harapan hidup tidak memiliki outlier sama sekali

Untuk menghindari penghapus data terlalu banyak berdasarkan outlier, maka akan dibandingkan terlebih dahulu dengan upper (Q3 + (1.5 * IQR)) dan lower (Q1 - (1.5 * IQR)) whiskernya. Jika hasil perbandingan dengan upper dan lower whiskernya masih terlalu banyak outlier, akan ditentukan dengna nilai ekstrim (Q3 + (3 * IQR) dan (Q1 - (3 * IQR)) dan nilai ekstrim itulah yang akan dihapus, bukan outliernya.

```python
#Mencari IQR
q1 = df[['population', 'life_exp', 'gdp_cap']].quantile(0.25)
q3 = df[['population', 'life_exp', 'gdp_cap']].quantile(0.75)
IQR = q3 - q1

#Mencari Upper dan Lower Whisker
uw = (q3 + (1.5 * IQR)) 
lw = (q1 - (1.5 * IQR))

#Mencari ekstrim upper dan lower
eu = (q3 + (3 * IQR))
el = (q1 - (3 * IQR))

#Gabungkan dalam bentuk dictionary
data = { 'nama_kolom': q1.index,
         'q1': q1.values,
         'q3': q3.values,
         'upper_whisker': uw.values,
         'lower_whisker': lw.values,
         'ekstrim_upper': eu.values,
         'ekstrim_lower': el.values
}

#Buat dalam bentuk dataframe
df_penghapus = pd.DataFrame(data = data)
df_penghapus
```

Selanjutnya, akan dilihat dibandingkan total outlier yang didapatkan jika menggunakan 
1. upper_whisker dan lower_whisker
2. Ekstrim_upper dan ekstrim_lower

```python
#Dengan upper_whisker dan lower_whisker
total_outlier_uw_lw = ((df[['population', 'life_exp', 'gdp_cap']] < lw) | (df[['population', 'life_exp', 'gdp_cap']] > uw)).sum()

#Dengan ekstrim_whisker dan lower_whisker
total_outlier_eu_el = ((df[['population', 'life_exp', 'gdp_cap']] < el) | (df[['population', 'life_exp', 'gdp_cap']] > eu)).sum()

#Menghitung proporsi persen
persen_uw_ul = (total_outlier_uw_lw / len(df)) * 100
persen_eu_el = (total_outlier_eu_el / len(df)) * 100

#Gabungkan data dalam dictionary
data = { 'nama_kolom': total_outlier_uw_lw.index,
         'Total_Upper_Lower_Whisker': total_outlier_uw_lw.values,
         'Total_Upper_Lower_Ekstrim': total_outlier_eu_el.values,
         'Persen(%)_Upper_Lower_Whisker': persen_uw_ul.values,
         'Persen(%)_Upper_Lower_Ekstrim': persen_eu_el.values,
    
}

#Masukkan dalam bentuk Data Frame
df_perbandingan_outlier = pd.DataFrame(data = data)
df_perbandingan_outlier
```

Hasil di atas dapat disimpulkan bahwa:
1. Total outlier menggunakan upper dan lower whisker memiliki total yang lebih banyak dikarenakan ketentuan angka untuk data outlier lebih kecil dibandingkan dengan upper dan lower ekstrim.
2. Sehingga dengan data yang dimiliki kurang 10000, maka data yang dinyatakan outlier akan menggunakan syarat upper dan lower ekstrim yang memiliki proporsi 8.86% dari seleruh data

Selanjutnya akan dilakukan penghapus menggunakan syarat upper dan lower ekstrim

```python
#Melakukan copy terlebih dahulu untuk tidak merusak data utama
df_hapus_outlier = df.copy()

df_hapus_outlier = df[~((df_hapus_outlier[['population', 'life_exp', 'gdp_cap']] < el) | (df_hapus_outlier[['population', 'life_exp', 'gdp_cap']] > eu)).any(axis = 1)]
df_hapus_outlier
```

Selanjutnya akan dicek berapa korelasi yang dihasilkan setelah penghapus outlier

```python
spearman_correlation(df_hapus_outlier)
```

Untuk memudahkan analisis perbandingan, mari kita buat hasilnya bersebelahan

```python
fig, axes = plt.subplots(1,2, figsize = (15, 5), sharex = True)

#Buat fungsi baru pada spearman korelasi perbandingan
def spearman_correlation_ax(df,kolom, judul):
    axes[kolom].set_title('Korelasi {}'.format(judul), fontsize = 15, pad = 15)
    ax = sns.heatmap(df.corr(method = 'spearman'), annot = True, annot_kws = {'fontsize' : 15}, linewidth = 0.1, ax = axes[kolom])
    axes[kolom].tick_params(axis = 'x', labelsize = 13)
    axes[kolom].tick_params(axis = 'y', labelsize = 13)


spearman_correlation_ax(df, 0, 'Sebelum Hapus Outlier')
spearman_correlation_ax(df_hapus_outlier, 1, 'Setelah Hapus Outlier')

plt.savefig('korelasi.png')
```

Hasil perbandingan didapatkan informasi bahwa Penghapusan outlier ternyata mengurangi korelasi antar variabel yang dibuktikan pada kolom life_expation dari 0.83 menjadi 0.82. Namun berbanding terbalik dengan korelasi antar populasi yang justru menaikkan dari 0.052 menjadi 0.072. Dikarenakan terjadi penurunan yang berbanding terbalik, mari kita kembali ke informasi sebelumnya tentang pengaruh populasi terhadap life_expation dan populasi dan life_expation terhadap GDP untuk menentukan data yang mana yang nantinya digunakan untuk melatih model.
